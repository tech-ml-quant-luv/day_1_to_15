#### Combining data into one file

```python

folder_path = "./5_min_combined_data/COMBINED_DATA"
column_to_extract = "close"

all_files = glob.glob(os.path.join(folder_path, "*.csv"))
print(all_files)

result_df = pd.DataFrame()

for file in all_files:
    base_name = os.path.basename(file).replace(".csv", "")
    print(base_name)

    # read datetime column properly
    df = pd.read_csv(file, parse_dates=["datetime"])
    df = df.set_index("datetime")

    # extract the close column
    col = df[[column_to_extract]].copy()

    # rename column: e.g., reliance_close
    new_col_name = f"{base_name}_{column_to_extract}"
    col.columns = [new_col_name]

    # merge aligned by datetime index
    result_df = pd.concat([result_df, col], axis=1)

# Save final dataframe
result_df.to_csv("./major_database/nifty_50_all_close_prices.csv")
```
