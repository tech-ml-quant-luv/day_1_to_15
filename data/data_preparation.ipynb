```python
import pandas as pd
import os

for symbol in nifty50_tickers:
    folder_and_file = symbol.replace("NSE:", "").replace("-EQ", "").lower()

    # Reset year & month for each symbol
    year = start_year
    month = start_month

    # List to collect dfs
    dfs = []

    while (year < end_year) or (year == end_year and month <= end_month):

        month_str = f"{month:02d}"

        file_path = f"./5_min_data/{folder_and_file}_eq/{folder_and_file}_eq_{year}_{month_str}.csv"

        if os.path.exists(file_path):
            df = pd.read_csv(file_path).set_index("datetime")
            dfs.append(df)
            print(f"Loaded: {file_path}")
        else:
            print(f"Missing file: {file_path}")

        # Move to next month
        if month == 12:
            month = 1
            year += 1
        else:
            month += 1

    # -----------------------------
    # Combine all dfs after while loop
    # -----------------------------
    if dfs:
        final_df = pd.concat(dfs).sort_index()

        # Ensure output folder exists
        os.makedirs("./5_min_combined_data/COMBINED_DATA", exist_ok=True)

        save_path = f"./5_min_combined_data/COMBINED_DATA/{folder_and_file}.csv"
        final_df.to_csv(save_path)

        print(f"\nSaved combined file: {save_path}\n")
    else:
        print(f"No data found for {folder_and_file}")
```
