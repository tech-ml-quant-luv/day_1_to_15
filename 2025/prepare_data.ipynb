```python
import pandas as pd
import glob
```

```python
def wrangle(file_path):
    df = pd.read_csv(file_path).set_index("Time")
    drop_columns = ["Ticker", "Date", "Volume", "Open Interest", "Open", "High", "Low"]
    df = df.drop(columns=drop_columns)
    
    return df
```

```python
files = sorted(glob.glob("./AUG/GFDLCM_STOCK_*.csv"))

all_dfs = []

for f in files:
    processed_df = wrangle(f)
    all_dfs.append(processed_df)

# Concatenate into one big dataframe
final_df = pd.concat(all_dfs, ignore_index=True)


print("\nFinal combined shape:", final_df.shape)

```

```python
final_df.to_csv("may_data.csv")
```
